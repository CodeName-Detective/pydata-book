# Figures and SubPlots

```python
fig = plt.figure() # Plots in matplotlib reside within a Figure Object
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(),'k--') #By default draws on the last subplot
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
```

### Method 2

```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha=0.5)

plt.subplots_adjust(wspace=0,hspace=0)
```

```python
plt.plot(np.random.randn(30).cumsum(),'ko--')
```

```python
plt.plot(np.random.randn(30).cumsum(),color='k',linestyle='dashed',marker='o')
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data,color='k',linestyle='dashed',marker='o',label='default')
plt.plot(data,color='k',linestyle='dashed',marker='o',drawstyle='steps-post',label='step-post')
plt.legend(loc='best')
```

# Ticks, Labels and Legends

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) # Create a axes object
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
ax.plot(np.random.randn(1000).cumsum())
```

```python
#Adding Legends
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),'k',label='one')
ax.plot(np.random.randn(1000).cumsum(),'k--',label='Two')
ax.plot(np.random.randn(1000).cumsum(),'k.', label='Three')
ax.legend(loc='best')
```

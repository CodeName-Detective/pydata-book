# Time Series with Duplicate Indices

```python
dates = pd.DatetimeIndex(['1/1/2000','1/2/2000','1/2/2000','1/2/2000','1/3/2000'])
dup_ts = pd.Series(np.arange(5), index=dates)
dup_ts
```

```python
dup_ts.index.is_unique
```

```python
dup_ts['1/03/2000']
```

```python
dup_ts['1/02/2000']
```

```python
grouped = dup_ts.groupby(level=0)
grouped.mean()
```

```python
grouped.count()
```
